# Filtering to Playlists modified in 2017
I ran into the GH large file limit of 100MB per file:
- playlists.csv = 1.69 GB
- unique_albums.csv = 43 MB
- unique_artists.csv = 15.5 MB
- unique_tracks = 400 MB
- all_zipped_together = 884 MB

## Filter to only playlists made in 2017

```python
start_2017 = 1483228800
end_2017 = 1514764799
playlist_df_17 = playlist_df[(start_2017 <= playlist_df.modified_at) & (playlist_df.modified_at <= end_2017)]
playlist_df_17
```

```python
# Convert modified_at column to readable date columns
playlist_df_17['modified_at_date'] = playlist_df_17.apply(lambda row: pd.Timestamp(row.modified_at, unit='s'), axis=1)

playlist_df_17['modified_at_year'] = playlist_df_17.apply(lambda row: row.modified_at_date.year, axis=1)
playlist_df_17['modified_at_month'] = playlist_df_17.apply(lambda row: row.modified_at_date.month, axis=1)
playlist_df_17['modified_at_day'] = playlist_df_17.apply(lambda row: row.modified_at_date.day, axis=1)
playlist_df_17['modified_at_dow'] = playlist_df_17.apply(lambda row: row.modified_at_date.dayofweek, axis=1)

playlist_df_17.head(5)
```

```python
# Save 2017 to csv
print('Saving to csv.')
playlist_df_17.to_csv(f"{output_filepath_root}/playlists_2017.csv", index=True)
print(f'Saved to {output_filepath_root}/playlists_2017.csv')
```

### Creating Unique Track DF from 2017 Playlists

```python
# Get track columns from playlist DF
track_uri_columns = [col for col in playlist_df_17.columns if col[:5] == 'track' and 'album' not in col and 'artist' not in col]
track_uri_columns
```

```python
track_df.index[:5]
```

```python
[t for t in playlist_df_17['track_1_uri'].unique() if t in track_df.index]
```

```python
# Get unique tracks in each of these track_#_uri columns

# list of lists
unique_track_17_uris_list = []
for col in track_uri_columns:
    this_col_uris = [t for t in playlist_df_17[col].unique() if t in track_df.index]
    unique_track_17_uris_list.append(this_col_uris)

unique_track_17_uris = []

for uri_list in unique_track_17_uris_list:
    for uri in uri_list:
        unique_track_17_uris.append(uri)

unique_track_17_uris[0]
unique_track_17_uris = set(unique_track_17_uris)
len(unique_track_17_uris)
```

Now that we have the unique URIs in the 2017 dataframe, grab rows from the big `track_df` that have URIs in this list

```python
track_df.index
```

```python
bool_mask = [uri in unique_track_17_uris for uri in track_df.index]
track_df_17 = track_df[bool_mask]

track_df_17
```

### Calculating Popularity of Tracks

```python
# Columns in playlist_df with track URIs
track_uri_cols = [col for col in playlist_df_17.columns
                  if 'uri' in col
                  and 'album' not in col
                  and 'artist' not in col]

# Use dictionary to hold song appearances
track_counts_dict = {}
for col in track_uri_cols:
    for uri in playlist_df_17[col].tolist():
        if uri not in track_counts_dict.keys():
            track_counts_dict.update({uri: 1})
        else:
            track_counts_dict.update({uri: track_counts_dict.get(uri) + 1})

track_counts_dict
```

```python
track_df_17.reset_index(inplace=True)
track_df_17['popularity'] = track_df_17.reset_index().apply(lambda row: int(track_counts_dict[row.track_uri]), axis=1)
track_df_17.set_index('track_uri', inplace=True)
track_df_17.head()
```

```python
# Save 2017 tracks to csv
print('Saving to csv.')
track_df_17.to_csv(f"{output_filepath_root}/tracks_2017.csv", index=True)
print(f'Saved to {output_filepath_root}/tracks_2017.csv')
```

### Creating Unique Albums & Artists from 2017 Unique Tracks

```python
unique_albums = track_df_17.reset_index().drop_duplicates(subset=['album_uri']).drop(columns=['track_name', 'track_uri', 'artist_uri', 'duration_ms'])
display(unique_albums)

unique_artists = track_df_17.reset_index().drop_duplicates(subset=['artist_uri']).drop(columns=['track_name', 'track_uri', 'album_name', 'album_uri', 'duration_ms'])
display(unique_artists)
```

```python
# Save these new DFs to csv and load them in subsequent notebook and convert to dictionaries
try:
    unique_albums.to_csv(f"{output_filepath_root}/albums_2017.csv", index=False)
    print(f"albums_2017.csv saved to {output_filepath_root}")
    unique_artists.to_csv(f"{output_filepath_root}/artists_2017.csv", index=False)
    print(f"artists_2017.csv saved to {output_filepath_root}")
except:
    print("CSV saving failed.")
```
